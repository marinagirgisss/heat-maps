```python
pr_coral_data = pd.read_csv("input/NCRMP_Puerto_Rico_corals.csv")
pr_coral_data = pr_coral_data.set_index("species_name")
pr_coral_data["Unknown_percent"] = pr_coral_data["unknown_count"]/pr_coral_data["total_diseased_count"]*100
pr_coral_data["Healthy_percent"] = pr_coral_data["healthy_count"]/pr_coral_data["total_diseased_count"]*100


pr_coral_data= pr_coral_data.drop("unknown_count",axis=1)
pr_coral_data= pr_coral_data.drop("healthy_count",axis=1)
pr_coral_data= pr_coral_data.drop("total_diseased_count",axis=1)

plot2 = sns.heatmap(pr_coral_data, annot=True, cmap='viridis')
plot2.set(title="Puerto Rico Coral Disease Data Heatmap", ylabel="Coral Species");
```

```python
persian_gulf_coral_data = pd.read_csv("input/Persian_gulf_Hazraty-Kari_et_al.csv", index_col=1)

# Remove weird first column and year
persian_gulf_coral_data = persian_gulf_coral_data.drop(persian_gulf_coral_data.columns[0], axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("year", axis=1)
persian_gulf_coral_data["BBD_percent"] = persian_gulf_coral_data["BBD_count"]/persian_gulf_coral_data["n"]
persian_gulf_coral_data["SGA_percent"] = persian_gulf_coral_data["SGA_count"]/persian_gulf_coral_data["n"]
persian_gulf_coral_data["WMD_percent"] = persian_gulf_coral_data["WMD_count"]/persian_gulf_coral_data["n"]
persian_gulf_coral_data["AYBD_percent"] = persian_gulf_coral_data["AYBD_count"]/persian_gulf_coral_data["n"]
persian_gulf_coral_data["WS_percent"] = persian_gulf_coral_data["WS_count"]/persian_gulf_coral_data["n"]
persian_gulf_coral_data["PPS_percent"] = persian_gulf_coral_data["PPS_count"]/persian_gulf_coral_data["n"]

persian_gulf_coral_data = persian_gulf_coral_data.drop("BBD_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("SGA_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("WMD_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("AYBD_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("WS_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("PPS_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("healthy_count", axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("total_diseased_count", axis=1)

persian_gulf_coral_data = persian_gulf_coral_data.drop("n", axis=1)



plot3 = sns.heatmap(persian_gulf_coral_data, annot=True, cmap='viridis')
plot3.set(title="Persian Gulf Coral Disease Data Heatmap", ylabel="Coral Species");
```

```python
isle_data = pd.read_csv("input/Sere_et_al_2015_Isle_de_la_Reunion_Disease_Data.csv", index_col=0)

# Remove region, reef part, and source columns
isle_data = isle_data.drop(["Region", "Reef_Part", "Source"], axis=1)

# Remove None and unknown values from table
isle_data_numeric = isle_data.apply(pd.to_numeric, errors='coerce')

plot3 = sns.heatmap(isle_data_numeric, annot=True, cmap='viridis')
plot3.set(title="Isle de la Reunion Coral Disease Data Heatmap", ylabel="Coral Species");
```
